### Hourly prices (past 2k hours)

```python
import requests
import pandas as pd
import json
res = requests.get('https://min-api.cryptocompare.com/data/histohour?fsym=BTC&tsym=USD&limit=2000&aggregate=1&e=CCCAGG')
btc = pd.DataFrame(json.loads(res.text)['Data'])
btc['coin'] = 'BTC'
res = requests.get('https://min-api.cryptocompare.com/data/histohour?fsym=ETH&tsym=USD&limit=2000&aggregate=1&e=CCCAGG')
eth = pd.DataFrame(json.loads(res.text)['Data'])
eth['coin'] = 'ETH'

df = btc.append(eth)
df.tail()
```
