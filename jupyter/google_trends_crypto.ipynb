```python
from datetime import date
from datetime import timedelta
from pytrends.request import TrendReq
import pandas as pd
import numpy as np

#Create req. container
trends= TrendReq(hl='en-US', tz=360)
keywords = ['bitcoin', 'ethereum']

#Get trends for past 180 * 12 days
d = date.today()

def get_prev_180_days(end_date):
    start_date = end_date - timedelta(days=180)
    timeframe = start_date.strftime('%Y-%m-%d') + ' ' + end_date.strftime('%Y-%m-%d')
    trends.build_payload(keywords, cat=0, timeframe=timeframe, geo='', gprop='')
    return trends.interest_over_time()

df = pd.DataFrame()
for i in range(0,14):
    df = df.append(get_prev_180_days(d))
    d = d - timedelta(days=180)
```

```python
#df['date'] = df.index
#df['date']=df['date'].apply(lambda x: x.strftime('%Y-%m-%d'))
df = df.iloc[::-1]
df.tail(10)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()
```

```python
x_vals = np.linspace(1,df.shape[0],num=df.shape[0])
plot = figure(x_axis_label = 'date', y_axis_label='interest (normalized)')
plot.line(x_vals, df['bitcoin'], legend="Bitcoin", color='blue')
plot.line(x_vals, df['ethereum'], legend="Ethereum", color='red')
show(plot)
```
