```python
import numpy as np
import pandas as pd
from pytrends.request import TrendReq

trends = TrendReq(hl='en-US', tz=360)
keyword = ["blockchain"]
trends.build_payload(keyword, cat=0, timeframe='today 5-y', geo='', gprop='')

blockchain_ot = trends.interest_over_time()

#First problem - dash + ripple signals will contain noise
keywords = ["bitcoin", "ethereum", "dash", "ripple", "litecoin"]
trends.build_payload(keywords, cat=0, timeframe='today 5-y', geo='', gprop='')

cryptos_ot = trends.interest_over_time()
```

```python
blockchain_ot.tail()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()
```

```python
#blockchain_ot
x_vals = np.linspace(1,259,num=259)
plot = figure(x_axis_label = 'week', y_axis_label='interest')
plot.line(x_vals, blockchain_ot['blockchain'])
show(plot)
```

```python
x_vals = np.linspace(1,259,num=259)
plot = figure(x_axis_label = 'week', y_axis_label='interest')
plot.line(x_vals, cryptos_ot['bitcoin'], legend="Bitcoin")
plot.line(x_vals, cryptos_ot['ethereum'], legend="Ethereum", line_color="red")
plot.line(x_vals, cryptos_ot['dash'], legend="Dash", line_color="purple")
plot.line(x_vals, cryptos_ot['ripple'], legend="Ripple", line_color='green')
plot.line(x_vals, cryptos_ot['litecoin'], legend="Litecoin", line_color='orange')
show(plot)
```
